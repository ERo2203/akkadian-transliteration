```python
import pandas as pd

df = pd.read_csv("../data/mtm24_transliterated.csv")

print(df.shape)
print(df.columns.tolist())
```

```python
print(df.isnull().sum())
```

```python
from collections import Counter

sign_counter = Counter()

for text in df["original_cuneiform"].astype(str):
    sign_counter.update(text)

print("Unique signs:", len(sign_counter))

print("\nTop 20 signs:")
for sign, freq in sign_counter.most_common(20):
    print(repr(sign), freq)
```

```python
char_counter = Counter()

for text in df["transliteration"].astype(str):
    char_counter.update(text)

print("Unique transliteration chars:")
print(len(char_counter))

print("\nCharacters:")
print(sorted(char_counter.keys()))
```

```python
import numpy as np

cuneiform_lengths = [
    len(str(x))
    for x in df["original_cuneiform"]
]

translit_lengths = [
    len(str(x))
    for x in df["transliteration"]
]

print("CUNEIFORM")

print("mean =", np.mean(cuneiform_lengths))
print("95th =", np.percentile(cuneiform_lengths,95))
print("max =", np.max(cuneiform_lengths))

print()

print("TRANSLITERATION")

print("mean =", np.mean(translit_lengths))
print("95th =", np.percentile(translit_lengths,95))
print("max =", np.max(translit_lengths))
```

```python
token_counter = Counter()

for text in df["transliteration"].astype(str):
    token_counter.update(text.split())

print(token_counter.most_common(50))
```
