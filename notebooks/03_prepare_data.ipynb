```python
import pandas as pd

df = pd.read_csv("../data/mtm24_transliterated.csv")

df = df[
    [
        "original_cuneiform",
        "transliteration"
    ]
]

df.head()
```

```python
import re

def normalize_translit(text):

    text = str(text)

    text = re.sub(r"\s+", " ", text)

    return text.strip()
```

```python
df["transliteration"] = (
    df["transliteration"]
    .apply(normalize_translit)
)
```

```python
src_chars = set()

for text in df["original_cuneiform"]:
    src_chars.update(text)

len(src_chars)
```

```python
tgt_chars = set()

for text in df["transliteration"]:
    tgt_chars.update(text)

len(tgt_chars)
```

```python
src_vocab = {
    "<pad>":0,
    "<sos>":1,
    "<eos>":2,
    "<unk>":3
}

for c in sorted(src_chars):
    src_vocab[c] = len(src_vocab)
```

```python
tgt_vocab = {
    "<pad>":0,
    "<sos>":1,
    "<eos>":2,
    "<unk>":3
}

for c in sorted(tgt_chars):
    tgt_vocab[c] = len(tgt_vocab)
```

```python
import json

with open(
    "../models/src_vocab.json",
    "w",
    encoding="utf8"
) as f:
    json.dump(
        src_vocab,
        f,
        ensure_ascii=False
    )

with open(
    "../models/tgt_vocab.json",
    "w",
    encoding="utf8"
) as f:
    json.dump(
        tgt_vocab,
        f,
        ensure_ascii=False
    )
```

```python
print("Source vocab:", len(src_vocab))
print("Target vocab:", len(tgt_vocab))
```
