```python
import re

def normalize_translit(text):
    text = str(text)

    text = re.sub(r"\s+", " ", text)

    return text.strip()
```

```python
import pandas as pd

df = pd.read_csv("../data/mtm24_transliterated.csv")

print(df.shape)
df.head()
```

```python
from collections import Counter

token_counter = Counter()

for text in df["transliteration"]:
    token_counter.update(text.split())

print("Unique tokens:", len(token_counter))
```

```python
print(token_counter.most_common(100))
```
